```python
### Practice with loops
## while loop
# not used often but can be very useful for numerical calculation
# essentially an if/elif/else statement looped
# while condition :
#   expression
error = 50
while error > 1 :
    error = error / 4
    print(error)
# notice here that the code will run until the error is less than 1 
```

```python
# more complex while loop that includes conditionals
offset = -6
while offset != 0 :
    print("connecting...")
    if offset > 0 :
        offset = offset - 1
    else :
        offset = offset + 1
    print(offset)
# this ensures that no matter if offset is a positive or negative number, the while loop will be able to complete without interruption
```

```python
## Now let's get into the for loop
# for var in seq :
# expression

# if I want to print out each of the values of fam:
fam = [1.73, 1.68,1.71, 1.89]
for height in fam : 
    print(height)
```

```python
# if we want to print out the index of the values of fam along with their respective value:
for index, height in enumerate(fam) :
    print("index" + str(index) + ": " + str(height))
```

```python
# for loops can be utilized for most data structures
# this includes strings:
for c in "family" :
    print(c.capitalize())
```

```python
# and lists of lists
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]

for room, size in house : 
    print("the " + str(room) + " is " + str(size) + " sqm")

# or
for x in house : 
    print("the " + x[0] + " is " + str(x[1]) + " sqm") # advantageous for lists of lists that have many subcategories
```

```python
# and numpy arrays (.nditer())
import numpy as np

np_height = np.array([1.73, 1.68, 1.71, 1.81, 1.79])
np_weight = np.array([65.4, 59.2, 63.6, 88.4, 68.7])
meas = np.array([np_height, np_weight])

for x in np.nditer(meas) : # one cannot simply say 'for x in meas' because they will produce entire arrays
    print(x)

```

```python
# and dictionaries (.items())
world = {"afghanistan": 30.55, "albania":2.77, "algeria":39.21}
for key, value in world.items():
    print(key + " -- " + str(value))
# the key here is .items()!!!
```

```python
# and finally pandas data frames (.iterrows())
import pandas as pd

wine = pd.read_csv("wine.csv")
for lab, row in wine.iterrows() : 
    print(lab)
    print(row)
```

```python
for lab, row in wine.iterrows() : 
    print(str(lab + 1) + " : " + "the pH is " + str(row['pH']))
```
