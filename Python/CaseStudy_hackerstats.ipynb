```python
### Case Study: Hacker Statistics
import pandas as pd
import numpy as np
```

```python
# generate a random number between 0 and 1 
# first want to set the seed
np.random.seed(123)
np.random.rand()
coin_toss = np.random.randint(0,2) # 2 because it is not inclusive
print(coin_toss)
```

```python
# now let's say you want to simulate a dice throw 50 times (empire state scenario)
np.random.seed(23)
step = 50
dice = np.random.randint(1,7)
if dice <= 2 :
    step = step - 1
elif dice <= 5 :
    step = step + 1
else :
    step = step + np.random.randint(1,7)

print(dice)
print(step)
```

```python
# simple for loop with lists to record all outcomes of head and tails
np.random.seed(69)
outcomes = []
for x in range(10) :
    coin = np.random.randint(0,2)
    if coin == 0 :
        outcomes.append('heads')
    else :
        outcomes.append('tails')
print(outcomes)   
```

```python
# how about ading the amount of tails that occur in the trial (like random walk)
np.random.seed(33)
tails = [0]
for x in range(10) :
    coin = np.random.randint(0,2)
    tails.append(tails[x] + coin)
print(tails) #shows that there were 5 tails thrown and exactly when they were thrown
```

```python
# now let's do an actual random walk
np.random.seed(123)
random_walk = [0]
for x in range(100) :
    step = random_walk[-1] # setting to last value of random_walk
    dice = np.random.randint(1,7)
    if dice <= 2 :
        step = max(0, step - 1) # because we can't go past the first step at the empire state building
    elif dice <= 5 :
        step = step + 1
    else :
        step = step + np.random.randint(1,7)
    random_walk.append(step) # append random_walk with the step value
print(random_walk)
```

```python
# let's visualize that random walk
import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.title('Random Walk: Empire State Scenario')
plt.xlabel('Throw #')
plt.ylabel('Step #')
plt.show()
```

```python
## Okay but what if we want to know the distribution of the last value of a scenario like this?
# for the heads/tails scenario, we can do this by:
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(11)
final_tails = []
for x in range(100):
    tails = [0]
    for x in range(10):
        coin = np.random.randint(0,2)
        tails.append(tails[x] + coin)
    final_tails.append(tails[-1])

plt.hist(final_tails, bins=6)
plt.title('Distribution of Tails over 10 throws 100 sims')
plt.xlabel('# of Tails in 10 throws')
plt.ylabel('Total # of tails through 100 sims')
plt.show()
```
